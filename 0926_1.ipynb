```python
import numpy as np

# 함수 정의
def f(x):
    return 6*x**2 + 7*x + 8



# 2. 적분 (리만 합)
def riemann_integral(f, a, b, n=10000):
    dx = (b - a) / n
    x = np.linspace(a, b-dx, n)  # 왼쪽 리만 합
    return np.sum(f(x) * dx)

# 테스트
a_point = 2
print("미분 (f'(2) 근사) =", derivative(f, a_point))

integral_val = riemann_integral(f, 0, 3, n=10000)
print("적분 (∫_0^3 f(x) dx 근사) =", integral_val)

# 검증 (정확 해석값)
# f'(x) = 12x + 7, f'(2)=31
print("실제 미분 f'(2) =", 12*a_point + 7)

# ∫_0^3 (6x^2+7x+8) dx = [2x^3 + (7/2)x^2 + 8x]_0^3
exact_integral = (2*3**3 + (7/2)*3**2 + 8*3) - 0
print("실제 적분 =", exact_integral)
```

```python
import sympy as sp

# 변수 정의
x = sp.Symbol('x')

# 함수 정의
f = 3*x**2 + 7*x + 9

# 미분
f_prime = sp.diff(f, x)

# 결과 출력
print("f(x) =", f)
print("f'(x) =", f_prime)
```

```python
import sympy as sp

# 변수 정의
x = sp.Symbol('x')

# 함수 정의
f = 3*x**2 + 7*x + 9

# 도함수 계산
f_prime = sp.diff(f, x)

# 평가할 x 값 리스트
x_values = [-1, 2, 3, 4, 5]

# 각 x 값에 대해 도함수 값 계산
for val in x_values:
    derivative_value = f_prime.subs(x, val)
    print(f"f'({val}) = {derivative_value}")
```

```python
import matplotlib.pyplot as plt

# x 값: -10부터 9까지
x = [i for i in range(-10, 10)]

# y 값: 각 x에 대해 f(x) = 3x^2 + 7x + 9 계산
y = [3*i**2 + 7*i + 9 for i in x]

# 그래프 그리기
plt.plot(x, y, label='f(x) = 3x² + 7x + 9', color='blue')

# 축 레이블, 제목, 그리드
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = 3x² + 7x + 9')
plt.grid(True)
plt.legend()

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# plt.rcParams['font.family'] = 'AppleGothic'  # macOS
plt.rcParams['axes.unicode_minus'] = False

# x 값
x = [i for i in range(-10, 10)]

# 원래 함수 f(x) = 3x² + 7x + 9
y_f = [3*i**2 + 7*i + 9 for i in x]

# 도함수 f'(x) = 6x + 7
y_df = [6*i + 7 for i in x]

# 그래프 그리기
plt.plot(x, y_f, label='f(x) = 3x² + 7x + 9', color='blue')
plt.plot(x, y_df, label="f'(x) = 6x + 7", color='red', linestyle='--')

# 한글 제목/라벨
plt.xlabel('x 값')
plt.ylabel('함수값')
plt.title('함수와 도함수 그래프')
plt.grid(True)
plt.legend()

# 그래프 출력
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import sin, pi, sqrt, N

# 단위원 그리기
circle = plt.Circle((0, 0), 1, color='lightgray', fill=False, linewidth=2)

fig, ax = plt.subplots(figsize=(6,6))
ax.add_artist(circle)

# 축 설정
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
ax.grid(True)

# 각도 리스트 (도 단위)
degrees = [30, 45, 60, 90]

for deg in degrees:
    # 라디안으로 변환
    rad = deg * np.pi / 180
    
    # x, y 좌표 (코사인, 사인)
    x = np.cos(rad)
    y = np.sin(rad)
    
    # 선 그리기 (원점 -> 원 위 점)
    ax.plot([0, x], [0, y], label=f'{deg}° (sin={round(y, 3)})')
    
    # 점 찍기
    ax.plot(x, y, 'o')
    
    # 삼각형의 높이 (sin값) 표시 (점에서 x축까지)
    ax.plot([x, x], [0, y], linestyle='dotted', color='gray')

# x축, y축 선
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# 범례
ax.legend()

plt.title('단위원 위 각도별 사인값 시각화')
plt.show()
```
