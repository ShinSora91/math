```python
import numpy as np
```

```python
def degToRad(i):
    return i * np.pi / 180
```

```python
degrees = [30,45,60,90]

radians = [d*np.pi / 180 for d in degrees]
radians
```

```python
def rotation_matrix(deg):
    theta = degToRad(deg)
    return np.array([
    [np.cos(degToRad(30)) , -np.sin(degToRad(30))],
    [np.sin(degToRad(30)) , np.cos(degToRad(30))]
])
```

```python
points = np.array([
    [1, 0],
    [0, 1],
    [-1, 0],
    [0, -1]
])
```

```python
import matplotlib.pyplot as plt
```

```python
# 회전 각도
angle = 30
R = rotation_matrix(angle)

# 회전 적용
rotated_points = np.dot(points, R.T)

# 시각화
plt.figure(figsize=(6,6))
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)

# 원래 점
plt.scatter(points[:,0], points[:,1], color='blue', label='original')
# 회전된 점
plt.scatter(rotated_points[:,0], rotated_points[:,1], color='red', label=f'rotated {angle}°')

# 연결선
for i in range(len(points)):
    plt.plot([points[i,0], rotated_points[i,0]], 
             [points[i,1], rotated_points[i,1]], 
             'k--', alpha=0.5)

plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```

```python
def toRot(i):
    theta = degToRad(i)
    return np.array([[np.cos(theta) , -np.sin(theta)],
            [np.sin(theta) , np.cos(theta)]])
```

```python
res = [toRot (i) for i in [30,45,60]]
res
```
