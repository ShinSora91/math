$f'(x) = \lim_{h\to 0}\frac{f(x+h)-f(x)}{h}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
# 대상 함수: f(x) = x^2
def f(v):
    return v * v
```

```python
x = np.linspace(-10, 10, 200)
y = f(x)
```

```python
def derivative(f, x):
    h= 0.00001
    return (f(x+h)-f(x))/h
```

```python
## 2) 수치 미분 (정의 사용)
y_der_num = [derivative(f, xi) for xi in x]
plt.plot(x, y, label='y = x^2')
plt.plot(x, y_der_num, label='dy/dx (numerical, central diff)')
```

```python
x = np.linspace(-10,10,100)
y_x_square_der = [derivative(f,j) for j in x]
plt.plot(x, x**2, label='x^2')
plt.plot(x, y_x_square_der, label='x^2_derivative')
```

```python
y_derivative = [derivative(np.cos,i) for i in np.linspace(-10,10,100)]
plt.plot(x, y, label = 'cos(x)')
plt.plot(x, y_derivative, label = '-sin(x)')
```
